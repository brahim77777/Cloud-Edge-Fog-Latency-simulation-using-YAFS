# ☁️🌫️📡 Cloud vs Fog vs Edge — Latency Simulation for IoT Smart Camera
### A YAFS Discrete-Event Simulation with Real Image Processing & Stochastic Network Noise

---

## Use Case & Motivation

A **smart surveillance camera** captures frames that must be processed (enhancement, filtering, edge detection) and the result returned within **1 second**. The key architectural question is:

> *Where should processing happen — Cloud, Fog, or Edge — to best meet the real-time constraint under realistic network conditions?*

### Network Architecture
```
  [Camera / Edge Node 0] ——3ms——▶ [Fog Node 1] ——201ms——▶ [Cloud Node 2]
         IPT = 100                   IPT = 1,000               IPT = 5,000
```

### Fundamental Computing Principle
Each tier has a **different compute capacity** measured in IPT (Instructions Per simulation Tick = per ms):

| Tier | IPT | Processing Time | Why |
|---|---|---|---|
| **Cloud** | 5,000 | Runtime-computed (fastest) | High-end datacenter server |
| **Fog** | 1,000 | Calibration reference | Local mid-range server |
| **Edge** | 100 | Runtime-computed (slowest) | Low-power IoT device |

The **same workload** (fixed instruction count) is sent to each tier. Since `processing_time = instructions / IPT`, a more powerful node finishes faster. This correctly models the Cloud < Fog < Edge compute ordering.

> **Important:** the exact processing latencies are **not fixed values in this notebook**. They are recomputed after calibration on the current machine each time the simulation runs, so only the ordering and formula are invariant.

### Network Specifications
| Parameter | Value | Constraint |
|---|---|---|
| Edge local dispatch | **0.042 ms** | ≤ 0.1 ms |
| Edge–Fog round-trip | **6 ms** | ≤ 6 ms |
| Edge–Cloud round-trip | **408 ms** | ≤ 800 ms |
| Real-time target | **1,000 ms** | Safety constraint |

### Expected Academic Narrative
- **Cloud:** Fastest compute, but the WAN round-trip is still significant, so end-to-end latency depends on the calibrated processing time plus the 408 ms network term.
- **Fog:** Medium compute with very small network delay, so it is often the best-balanced option.
- **Edge:** Zero network delay, but for this workload it is typically dominated by slower local compute.

This reveals the **fundamental tradeoff**: Cloud is network-bottlenecked, Edge is compute-bottlenecked, and Fog is the optimal balance point.

## 🖼️ Step 1: Image Acquisition & CPU Workload Characterization

### Purpose
We download a real camera-like image and apply a multi-pass image processing pipeline to:
1. **Generate a realistic workload** (contrast enhancement + edge-gradient transform)
2. **Measure actual CPU processing time** on this machine
3. **Derive the YAFS instruction count** that represents this workload
4. **Display before/after images** to visually confirm the processing

### Calibration Strategy
We first run a short calibration (10 iterations) to measure ms_per_iteration on this specific machine. We then compute N_ITERS so the full run targets the Fog reference budget on that machine. This value, translated to YAFS instructions, becomes the fixed workload sent to all three tiers.

Because this step depends on the measured calibration result, the resulting instruction count and all derived processing latencies can change from one run to another.

### Why Fog Is the Reference?
The Fog node is chosen as the calibration reference because it represents the **middle-tier baseline** — the most common deployment target for edge-cloud offloading. After calibration, Cloud processes the same instruction count faster and Edge processes it slower according to the IPT ratios; the exact milliseconds are produced at runtime.

````python
# ── Step 1A: Download image ─────────────────────────────────────────────────
IMG_URL = "https://upload.wikimedia.org/wikipedia/commons/4/47/PNG_transparency_demonstration_1.png"

print("[IMAGE] Downloading camera frame...")
resp = requests.get(IMG_URL, headers={'User-Agent': 'Mozilla/5.0'}, timeout=30)
img_original = Image.open(BytesIO(resp.content)).convert('RGB')
W, H = img_original.size
img_arr = np.array(img_original, dtype=np.float32)
print(f"[IMAGE] Loaded: {W}×{H} pixels  |  {W*H*3/1024:.1f} KB RGB")

# ── Step 1B: Define processing pipeline ────────────────────────────────────
def run_transform(arr, n_iters):
    """Image processing pipeline — n_iters passes over the frame.
    Each pass applies:
      1. Intensity transform  : gamma-like correction (sqrt + scale)
      2. Horizontal gradient  : Sobel-like edge enhancement
      3. Clip to valid range  : [0, 255]
    Mimics a lightweight pre-processing stage before AI inference.
    """
    out = arr.copy()
    for _ in range(n_iters):
        out = (np.sqrt(np.abs(out) + 1.0) * 16.0 + out * 0.5) % 256.0
        out[:, 1:, :] += 0.01 * (out[:, :-1, :] - out[:, 1:, :])
        out = np.clip(out, 0, 255)
    return out

# ── Step 1C: Calibration — measure ms per iteration ────────────────────────
CALIB_ITERS     = 10
FOG_TARGET_MS   = 900.0   # Fog reference processing time

print(f"\n[CALIB] Timing {CALIB_ITERS} iterations...")
t0 = time.perf_counter()
run_transform(img_arr, CALIB_ITERS)
ms_per_iter = (time.perf_counter() - t0) / CALIB_ITERS * 1000
N_ITERS = max(1, round(FOG_TARGET_MS / ms_per_iter))
print(f"[CALIB] {ms_per_iter:.2f} ms/iter → N_ITERS = {N_ITERS} to reach ~{FOG_TARGET_MS:.0f}ms")

# ── Step 1D: Run full processing & measure actual time ─────────────────────
print(f"\n[PROCESS] Running {N_ITERS} iterations on {W}×{H} image...")
t0 = time.perf_counter()
img_processed_arr = run_transform(img_arr, N_ITERS)
FOG_MEASURED_MS = (time.perf_counter() - t0) * 1000
img_processed = Image.fromarray(np.uint8(np.clip(img_processed_arr, 0, 255)))

print(f"[PROCESS] Measured time: {FOG_MEASURED_MS:.2f} ms  (Fog reference)")
print(f"[PROCESS] Calibration accuracy: ±{abs(FOG_MEASURED_MS-FOG_TARGET_MS)/FOG_TARGET_MS*100:.1f}%")

# ── Step 1E: Display before / after ────────────────────────────────────────
BG_DARK  = '#0f1117'
BG_PANEL = '#1a1d2e'

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.patch.set_facecolor(BG_DARK)
for ax, im, title, note in [
    (axes[0], img_original,
     'Original Camera Frame',
     f'{W}×{H} px  |  {W*H*3/1024:.0f} KB RGB'),
    (axes[1], img_processed,
     'Processed Output',
     f'{N_ITERS} transform passes  |  {FOG_MEASURED_MS:.0f} ms (Fog reference)'),
]:
    ax.imshow(im)
    ax.set_title(title, color='white', fontsize=13, pad=8)
    ax.set_xlabel(note, color='#aaa', fontsize=10)
    ax.axis('off')

plt.suptitle(
    f'IoT Smart Camera — Image Processing Workload\n'
    f'Fog reference: {FOG_MEASURED_MS:.0f} ms  |  '
    f'Cloud (5× faster IPT): ~{FOG_MEASURED_MS/5:.0f} ms  |  '
    f'Edge (10× slower IPT): ~{FOG_MEASURED_MS*10:.0f} ms',
    color='white', fontsize=12, fontweight='bold'
)
plt.tight_layout()
plt.show()

print(f"\n[OK] Workload characterized. FOG_MEASURED_MS = {FOG_MEASURED_MS:.2f} ms")
````

## 🔬 Step 2: All Constants — Defined Once

### Key Design: Fixed Instructions, Variable Processing Time

The workload is defined as a **fixed number of instructions** (derived from the current Fog measurement). Each tier then processes this workload at its own IPT rate:

```
INSTRUCTIONS = FOG_MEASURED_MS × FOG_IPT

Cloud processing = INSTRUCTIONS / 5,000   (fastest)
Fog   processing = INSTRUCTIONS / 1,000   (reference)
Edge  processing = INSTRUCTIONS /   100   (slowest)
```

This is the **physically correct model**: the same task takes longer on weaker hardware. The exact millisecond values shown later are runtime-calibrated outputs, not fixed assumptions.

````python
# ══════════════════════════════════════════════════════════════════════
#  ALL CONSTANTS — edit only here
# ══════════════════════════════════════════════════════════════════════

RANDOM_SEED           = 42
np.random.seed(RANDOM_SEED)

# ── Real-time constraint ───────────────────────────────────────────────
REALTIME_TARGET_MS    = 1_000.0

# ── Node compute capacities (IPT = instructions processed per ms) ──────
CLOUD_IPT             = 5_000
FOG_IPT               = 1_000
EDGE_IPT              =   100

# ── Fixed instruction count (derived from Fog measurement) ─────────────
# Fog takes FOG_MEASURED_MS to process this workload at FOG_IPT.
# Formula: instructions = measured_ms × IPT
INSTRUCTIONS          = FOG_MEASURED_MS * FOG_IPT   # e.g. ~900,000

# ── Processing time per tier (computed, not assumed) ───────────────────
# Formula: processing_ms = INSTRUCTIONS / node_IPT
CLOUD_PROC_MS         = INSTRUCTIONS / CLOUD_IPT    # ~180 ms  (fastest)
FOG_PROC_MS           = INSTRUCTIONS / FOG_IPT      # ~900 ms  (reference)
EDGE_PROC_MS          = INSTRUCTIONS / EDGE_IPT     # ~9,000 ms (slowest)

# ── Network one-way delays (ms) ────────────────────────────────────────
EDGE_LOCAL_MS         =   0.042   # local dispatch (≤ 0.1ms)
EDGE_FOG_ONE_WAY_MS   =   3.0     # LAN / 5G local link
FOG_CLOUD_ONE_WAY_MS  = 201.0     # regional WAN

# ── Derived round-trip times ───────────────────────────────────────────
EDGE_FOG_RT_MS        = EDGE_FOG_ONE_WAY_MS * 2
EDGE_CLOUD_RT_MS      = (EDGE_FOG_ONE_WAY_MS + FOG_CLOUD_ONE_WAY_MS) * 2

# ── Expected E2E delays per scenario (nominal, no noise) ───────────────
CLOUD_EXPECTED_E2E    = EDGE_CLOUD_RT_MS  + CLOUD_PROC_MS
FOG_EXPECTED_E2E      = EDGE_FOG_RT_MS    + FOG_PROC_MS
EDGE_EXPECTED_E2E     = EDGE_LOCAL_MS     + EDGE_PROC_MS

# ── Simulation parameters ──────────────────────────────────────────────
N_REQUESTS            = 30
SOURCE_EMIT_MS        = 1_000

# ── Network noise parameters ───────────────────────────────────────────
POISSON_LAMBDA_MS     = 2.0
NORMAL_SIGMA_MS       = 1.0

# Pre-generate noise arrays (shape: N_REQUESTS × 2 links)
np.random.seed(RANDOM_SEED)
poisson_noise = np.random.poisson(lam=POISSON_LAMBDA_MS, size=(N_REQUESTS, 2)).astype(float)
normal_noise  = np.abs(np.random.normal(loc=0, scale=NORMAL_SIGMA_MS, size=(N_REQUESTS, 2)))

# ── Verification ───────────────────────────────────────────────────────
print("═" * 68)
print("  WORKLOAD & PARAMETER SUMMARY")
print("═" * 68)
print(f"  Fixed instructions:    {INSTRUCTIONS:>12,.0f}")
print()
print(f"  {'Tier':<8} {'IPT':>6}  {'Processing':>14}  {'Ordering'}")
print(f"  {'-'*50}")
print(f"  {'Cloud':<8} {CLOUD_IPT:>6,}  {CLOUD_PROC_MS:>11.1f} ms  ← fastest (strongest)")
print(f"  {'Fog':<8} {FOG_IPT:>6,}  {FOG_PROC_MS:>11.1f} ms  ← reference")
print(f"  {'Edge':<8} {EDGE_IPT:>6,}  {EDGE_PROC_MS:>11.1f} ms  ← slowest (weakest)")
print()
assert CLOUD_PROC_MS < FOG_PROC_MS < EDGE_PROC_MS, "Processing order violated!"
print(f"  ✓ Order confirmed: Cloud ({CLOUD_PROC_MS:.0f}ms) < Fog ({FOG_PROC_MS:.0f}ms) < Edge ({EDGE_PROC_MS:.0f}ms)")
print()
print(f"  {'Scenario':<8} {'Network RT':>12}  {'Processing':>12}  {'E2E':>10}  {'vs RT'}")
print(f"  {'-'*60}")
for name, net, proc, e2e in [
    ('Cloud', EDGE_CLOUD_RT_MS, CLOUD_PROC_MS, CLOUD_EXPECTED_E2E),
    ('Fog',   EDGE_FOG_RT_MS,   FOG_PROC_MS,   FOG_EXPECTED_E2E),
    ('Edge',  EDGE_LOCAL_MS,    EDGE_PROC_MS,  EDGE_EXPECTED_E2E),
]:
    status = 'PASSES ✓' if e2e < REALTIME_TARGET_MS else f'FAILS ✗  ({e2e/REALTIME_TARGET_MS:.1f}×)'
    print(f"  {name:<8} {net:>10.1f}ms  {proc:>10.1f}ms  {e2e:>8.1f}ms  {status}")
print()
print(f"  Network constraints:")
print(f"  Edge local  {EDGE_LOCAL_MS:.3f}ms ≤ 0.1ms  ✓")
print(f"  Edge–Fog RT {EDGE_FOG_RT_MS:.1f}ms ≤ 6ms    ✓")
print(f"  Edge–Cloud  {EDGE_CLOUD_RT_MS:.0f}ms ≤ 800ms  ✓")
print("═" * 68)
````

## 📡 Step 3: Network Noise Models

Two stochastic distributions model realistic network perturbations:

**Loi de Poisson (λ = 2 ms)** — Bursty congestion: models packet arrival processes in a shared network (M/M/1 queue theory). Always non-negative, right-skewed tail represents occasional severe congestion bursts.

**Loi Normale clippée (μ = 0, σ = 1 ms)** — Symmetric jitter: models router processing time variation and transmission fluctuations (ITU-T Y.1541 jitter model). Clipped to [0, +∞) since delay cannot be negative.

Noise is injected **per-request** by rebuilding the YAFS topology with perturbed link delays for each of the 30 simulation runs.

````python
# ── Noise Model Visualization ───────────────────────────────────────────────
COLORS = {
    'Cloud': '#4472C4', 'Fog': '#ED7D31', 'Edge': '#70AD47',
    'Net':   '#5B9BD5', 'Proc': '#C00000',
    'Poi':   '#F4D03F', 'Nor':  '#85C1E9',
}

np.random.seed(RANDOM_SEED)
lp = np.random.poisson(lam=POISSON_LAMBDA_MS, size=10_000).astype(float)
ln = np.abs(np.random.normal(loc=0, scale=NORMAL_SIGMA_MS, size=10_000))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.patch.set_facecolor(BG_DARK)
for ax, data, color, title, subtitle in [
    (axes[0], lp, COLORS['Poi'],
     f'Loi de Poisson  (λ = {POISSON_LAMBDA_MS} ms)',
     'Bursty congestion — M/M/1 packet arrival model'),
    (axes[1], ln, COLORS['Nor'],
     f'Loi Normale clippée  (σ = {NORMAL_SIGMA_MS} ms)',
     'Symmetric jitter — ITU-T Y.1541 model'),
]:
    ax.set_facecolor(BG_PANEL)
    ax.hist(data, bins=40, color=color, alpha=0.85,
            edgecolor='white', linewidth=0.3, density=True)
    ax.axvline(data.mean(), color='white', lw=2, ls='--',
               label=f'Mean = {data.mean():.2f} ms')
    ax.axvline(data.mean()+data.std(), color='orange', lw=1.5, ls=':',
               label=f'Mean+σ = {data.mean()+data.std():.2f} ms')
    ax.set_title(title, color='white', fontsize=12, pad=8)
    ax.set_xlabel(subtitle, color='#aaa', fontsize=9)
    ax.set_ylabel('Probability Density', color='white')
    ax.tick_params(colors='white'); ax.spines[:].set_color('#444')
    ax.legend(facecolor=BG_PANEL, labelcolor='white', fontsize=9)
    ax.text(0.97, 0.97,
            f"mean = {data.mean():.2f} ms\nstd  = {data.std():.2f} ms\nP95  = {np.percentile(data,95):.2f} ms",
            transform=ax.transAxes, va='top', ha='right',
            color='white', fontsize=9,
            bbox=dict(boxstyle='round', fc='#2a2d3e', alpha=0.85))
plt.suptitle('Network Noise Models — 10,000 Samples', color='white',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
print("[OK] Noise models visualized.")
````

## 🗺️ Step 4: Shared Simulation Infrastructure

All three scenarios share the same topology and application structure. The **only variable** between scenarios is which node the `Process` module is deployed on — and the resulting processing time automatically differs because each node has a different IPT.

````python
# ── Shared Simulation Infrastructure ───────────────────────────────────────

def build_topology(ef_pr, fc_pr):
    """3-node directed topology. ef_pr: Edge-Fog delay (ms). fc_pr: Fog-Cloud delay (ms)."""
    G = nx.DiGraph()
    G.add_node(0, model='edge',  IPT=EDGE_IPT)
    G.add_node(1, model='fog',   IPT=FOG_IPT)
    G.add_node(2, model='cloud', IPT=CLOUD_IPT)
    G.add_edge(0, 1, BW=1000, PR=max(0.001, ef_pr))
    G.add_edge(1, 0, BW=1000, PR=max(0.001, ef_pr))
    G.add_edge(1, 2, BW=100,  PR=max(0.001, fc_pr))
    G.add_edge(2, 1, BW=100,  PR=max(0.001, fc_pr))
    topo = Topology()
    topo.G = topo.Gnx = G
    return topo


def build_application():
    """Source → Process (image task, INSTRUCTIONS fixed) → Sink.
    YAFS will compute processing_time = INSTRUCTIONS / node_IPT automatically.
    """
    app = Application(name='CameraApp')
    app.set_modules([
        {'Source':  {'Type': Application.TYPE_SOURCE}},
        {'Process': {'Type': Application.TYPE_MODULE}},
        {'Sink':    {'Type': Application.TYPE_SINK}},
    ])
    m1 = Message('M1', 'Source', 'Process', instructions=INSTRUCTIONS, bytes=50_000)
    m2 = Message('M2', 'Process', 'Sink',   instructions=100,          bytes=500)
    app.add_source_messages(m1)
    app.add_service_module('Process', m1, m2, lambda *a, **k: 1.0)
    return app


def run_one_request(placement_cls, results_dir, ef_pr, fc_pr, expected_e2e):
    """Run a single YAFS simulation. Returns E2E delay (ms) or None on failure."""
    os.makedirs(results_dir, exist_ok=True)
    os.chdir(results_dir)
    topo = build_topology(ef_pr, fc_pr)
    app  = build_application()
    dur  = int(SOURCE_EMIT_MS + expected_e2e * 1.3 + 5_000)
    sim  = Sim(topo)
    sim.deploy_app(app, placement_cls(name='P'), First_ShortestPath())
    sim.deploy_source(app.name, 0, app.get_message('M1'),
                      deterministic_distribution(name='Once', time=SOURCE_EMIT_MS))
    sim.deploy_sink(app.name, 0, 'Sink')
    sim.run(until=dur)
    try:
        df = pd.read_csv(os.path.join(results_dir, 'result.csv'))
        df.columns = df.columns.str.strip()
        rows = df[df['message'] == 'M2']
        if rows.empty:
            return None
        return float(rows.iloc[0]['time_reception']) - SOURCE_EMIT_MS
    except FileNotFoundError:
        return None


def run_scenario(name, placement_cls, base_ef, base_fc,
                 expected_e2e, expected_proc_ms, noise_arr):
    """Run N_REQUESTS simulations with per-request noise. Returns statistics dict."""
    results_dir = f'/content/sim_{name}'
    samples, failed = [], 0
    print(f'  [{name}] Running {N_REQUESTS} requests...', end=' ', flush=True)
    for i in range(N_REQUESTS):
        ef  = base_ef + float(noise_arr[i, 0])
        fc  = base_fc + float(noise_arr[i, 1])
        e2e = run_one_request(placement_cls, results_dir, ef, fc,
                              expected_e2e + float(noise_arr[i].sum()))
        if e2e and e2e > 0:
            samples.append(e2e)
        else:
            failed += 1
    print(f'Done. {len(samples)}/{N_REQUESTS} received.')
    if len(samples) < 5:
        raise RuntimeError(f'{name}: too few results ({len(samples)}).')
    s       = np.array(samples)
    mean_ms = float(np.mean(s))
    std_ms  = float(np.std(s, ddof=1))
    net_ms  = max(0.0, mean_ms - expected_proc_ms)
    ci95    = stats.t.interval(0.95, df=len(s)-1, loc=mean_ms, scale=stats.sem(s))
    return {
        'samples':        s,
        'n':              len(s),
        'mean_ms':        mean_ms,
        'std_ms':         std_ms,
        'p95_ms':         float(np.percentile(s, 95)),
        'p99_ms':         float(np.percentile(s, 99)),
        'net_ms':         net_ms,
        'proc_ms':        expected_proc_ms,
        'ci95':           ci95,
        'failure_factor': mean_ms / REALTIME_TARGET_MS,
        'meets_rt':       mean_ms < REALTIME_TARGET_MS,
        'violations':     int(np.sum(s >= REALTIME_TARGET_MS)),
        'violation_pct':  float(np.mean(s >= REALTIME_TARGET_MS)) * 100,
    }


# Sanity check
t = build_topology(EDGE_FOG_ONE_WAY_MS, FOG_CLOUD_ONE_WAY_MS)
print('[OK] Topology:', {n: d for n, d in t.G.nodes(data=True)})
print('[OK] Simulation infrastructure ready.')
````

# 🌩️ PART 1 — Cloud-Only Scenario

## Hypothesis
The Cloud node (IPT=5,000) processes the workload in the **smallest runtime-computed processing time** of the three tiers. With the current constants, the nominal network round-trip to Cloud is **408 ms**, so Cloud end-to-end latency is the calibrated processing time plus that WAN delay.

## Deployment
```
Camera(0) ——3ms——▶ Fog(1) ——201ms——▶ Cloud(2) [PROCESS — IPT=5,000]
Camera(0) ◀——3ms—— Fog(1) ◀——201ms——
Network RT = 408ms  |  Processing = runtime-computed  |  Expected E2E = network RT + processing
```

> **Note:** Cloud has the fastest compute, but the WAN latency still contributes a substantial fixed cost compared with Fog.

````python
# ── Cloud Placement ─────────────────────────────────────────────────────────
class CloudPlacement(Placement):
    """Deploy Process on Node 2 (Cloud, IPT=5,000 → fastest compute)."""
    def initial_allocation(self, sim, app_name):
        sim.deploy_module(app_name, 'Process',
                          sim.apps[app_name].services['Process'], [2])

print("[PART 1] CLOUD — IPT=5,000 → Processing ~{:.0f}ms".format(CLOUD_PROC_MS))

m_cloud_poi = run_scenario(
    'cloud_poi', CloudPlacement,
    EDGE_FOG_ONE_WAY_MS, FOG_CLOUD_ONE_WAY_MS,
    CLOUD_EXPECTED_E2E, CLOUD_PROC_MS, poisson_noise
)
m_cloud_nor = run_scenario(
    'cloud_nor', CloudPlacement,
    EDGE_FOG_ONE_WAY_MS, FOG_CLOUD_ONE_WAY_MS,
    CLOUD_EXPECTED_E2E, CLOUD_PROC_MS, normal_noise
)

for noise_label, m in [('Poisson', m_cloud_poi), ('Normal', m_cloud_nor)]:
    print(f'\n  ── Cloud / {noise_label} ─────────────────────────────────────')
    print(f'  IPT:                {CLOUD_IPT:,}  (strongest node)')
    print(f'  Processing time:    {CLOUD_PROC_MS:.1f} ms  (fastest: {INSTRUCTIONS:.0f} instr / {CLOUD_IPT} IPT)')
    print(f'  Network RT:         {m["net_ms"]:.1f} ms  (dominant component)')
    print(f'  E2E mean ± std:     {m["mean_ms"]:.2f} ± {m["std_ms"]:.2f} ms')
    print(f'  95% CI on mean:     [{m["ci95"][0]:.1f}, {m["ci95"][1]:.1f}] ms')
    print(f'  P95 / P99:          {m["p95_ms"]:.1f} ms / {m["p99_ms"]:.1f} ms')
    print(f'  RT violations:      {m["violations"]}/{m["n"]}  ({m["violation_pct"]:.0f}%)')
    print(f'  Failure factor:     {m["failure_factor"]:.4f}×')
    print(f'  Meets RT (mean):    {"YES ✓" if m["meets_rt"] else "NO  ✗"}')
````

### Part 1 — Academic Conclusion

Despite having the **fastest compute** after calibration, the Cloud scenario still pays the largest network cost. Under the current constants, the dominant issue is the **408 ms round-trip**, so Cloud performance depends on whether the calibrated processing time leaves enough headroom within the 1-second budget. This demonstrates that **raw compute speed does not guarantee the best end-to-end latency** when network distance is higher than Fog.

# 🌫️ PART 2 — Fog Computing Scenario

## Hypothesis
The Fog node (IPT=1,000) processes the workload in the **reference runtime-computed duration**. The 6 ms round-trip is negligible, so the expected E2E is the calibrated processing time plus a very small network term. This is typically more robust to noise than Cloud, despite slower compute.

## Deployment
```
Camera(0) ——3ms——▶ Fog(1) [PROCESS — IPT=1,000]
Camera(0) ◀——3ms——
Network RT = 6ms  |  Processing = runtime-computed  |  Expected E2E = network RT + processing
```

> **Key insight:** Fog is slower than Cloud in compute, yet often more reliable in meeting RT because its network path is dramatically shorter.

````python
# ── Fog Placement ───────────────────────────────────────────────────────────
class FogPlacement(Placement):
    """Deploy Process on Node 1 (Fog, IPT=1,000 → medium compute)."""
    def initial_allocation(self, sim, app_name):
        sim.deploy_module(app_name, 'Process',
                          sim.apps[app_name].services['Process'], [1])

# Fog only traverses Edge-Fog link — zero out Fog-Cloud noise
fog_poi = poisson_noise.copy(); fog_poi[:, 1] = 0.0
fog_nor = normal_noise.copy();  fog_nor[:, 1] = 0.0

print("[PART 2] FOG — IPT=1,000 → Processing ~{:.0f}ms".format(FOG_PROC_MS))

m_fog_poi = run_scenario(
    'fog_poi', FogPlacement,
    EDGE_FOG_ONE_WAY_MS, FOG_CLOUD_ONE_WAY_MS,
    FOG_EXPECTED_E2E, FOG_PROC_MS, fog_poi
)
m_fog_nor = run_scenario(
    'fog_nor', FogPlacement,
    EDGE_FOG_ONE_WAY_MS, FOG_CLOUD_ONE_WAY_MS,
    FOG_EXPECTED_E2E, FOG_PROC_MS, fog_nor
)

for noise_label, m in [('Poisson', m_fog_poi), ('Normal', m_fog_nor)]:
    print(f'\n  ── Fog / {noise_label} ───────────────────────────────────────')
    print(f'  IPT:                {FOG_IPT:,}  (medium node)')
    print(f'  Processing time:    {FOG_PROC_MS:.1f} ms  (reference: {INSTRUCTIONS:.0f} instr / {FOG_IPT} IPT)')
    print(f'  Network RT:         {m["net_ms"]:.2f} ms  (negligible)')
    print(f'  E2E mean ± std:     {m["mean_ms"]:.2f} ± {m["std_ms"]:.2f} ms')
    print(f'  95% CI on mean:     [{m["ci95"][0]:.1f}, {m["ci95"][1]:.1f}] ms')
    print(f'  P95 / P99:          {m["p95_ms"]:.1f} ms / {m["p99_ms"]:.1f} ms')
    print(f'  RT violations:      {m["violations"]}/{m["n"]}  ({m["violation_pct"]:.0f}%)')
    print(f'  Failure factor:     {m["failure_factor"]:.4f}×')
    print(f'  Meets RT (mean):    {"YES ✓" if m["meets_rt"] else "NO  ✗"}')
    print(f'  RT headroom (mean): {REALTIME_TARGET_MS - m["mean_ms"]:.1f} ms')
````

### Part 2 — Academic Conclusion

The Fog scenario often achieves the best balance: although it is slower in compute than Cloud, it can outperform Cloud in RT compliance because its network overhead is much smaller. The exact headroom depends on the current calibration, but Fog typically remains the most balanced single-tier architecture for this workload.

# 📡 PART 3 — Edge Computing Scenario

## Hypothesis
The Edge node (IPT=100) processes the workload in the **largest runtime-computed processing time** because it has the lowest IPT. With near-zero network overhead, the E2E is almost entirely compute-bound. For this workload, that makes Edge the worst-performing scenario despite having minimal network latency.

## Deployment
```
Camera(0) [SOURCE → PROCESS — IPT=100 → SINK]
Local dispatch = 0.042ms  |  Processing = runtime-computed  |  Expected E2E = local dispatch + processing
```

> **Key insight:** Zero network latency does not help when the device is compute-limited. Edge is appropriate only for **lightweight** tasks whose runtime-calibrated processing stays well below the RT target.

````python
# ── Edge Placement ──────────────────────────────────────────────────────────
class EdgePlacement(Placement):
    """Deploy Process on Node 0 (Edge, IPT=100 → slowest compute)."""
    def initial_allocation(self, sim, app_name):
        sim.deploy_module(app_name, 'Process',
                          sim.apps[app_name].services['Process'], [0])

# Edge: only local dispatch jitter (tiny, ~0.01ms sigma)
np.random.seed(RANDOM_SEED + 1)
local_jitter = np.abs(np.random.normal(0, 0.01, size=(N_REQUESTS, 2)))

print("[PART 3] EDGE — IPT=100 → Processing ~{:.0f}ms".format(EDGE_PROC_MS))

m_edge = run_scenario(
    'edge', EdgePlacement,
    EDGE_LOCAL_MS, FOG_CLOUD_ONE_WAY_MS,
    EDGE_EXPECTED_E2E, EDGE_PROC_MS, local_jitter
)

m = m_edge
print(f'\n  ── Edge / Local ───────────────────────────────────────────')
print(f'  IPT:                {EDGE_IPT:,}  (weakest node)')
print(f'  Processing time:    {EDGE_PROC_MS:.1f} ms  (slowest: {INSTRUCTIONS:.0f} instr / {EDGE_IPT} IPT)')
print(f'  Network RT:         {EDGE_LOCAL_MS:.3f} ms  (fully local, negligible)')
print(f'  E2E mean ± std:     {m["mean_ms"]:.1f} ± {m["std_ms"]:.1f} ms')
print(f'  95% CI on mean:     [{m["ci95"][0]:.1f}, {m["ci95"][1]:.1f}] ms')
print(f'  P95 / P99:          {m["p95_ms"]:.1f} ms / {m["p99_ms"]:.1f} ms')
print(f'  RT violations:      {m["violations"]}/{m["n"]}  ({m["violation_pct"]:.0f}%)')
print(f'  Failure factor:     {m["failure_factor"]:.2f}×  (worst performer)')
print(f'  Meets RT (mean):    {"YES ✓" if m["meets_rt"] else "NO  ✗"}')
````

### Part 3 — Academic Conclusion

Edge computing achieves **zero network latency** but remains **compute-bottlenecked** for this workload because the low-power IoT device has the smallest IPT. The exact processing time changes with calibration, but the conclusion does not: for a compute-intensive workload, Edge is unsuitable unless the task is lightweight enough to keep runtime-calibrated processing comfortably within the RT budget.
